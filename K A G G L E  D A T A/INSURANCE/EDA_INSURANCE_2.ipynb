# Incident Type vs Collision Type

```python
data['incident_type'].unique()
```

```python
data['incident_severity'].unique() 
```

```python
data['collision_type'].unique()
```

```python
incident = pd.crosstab(data['incident_type'], data['collision_type'])
colors = plt.cm.inferno(np.linspace(0, 1, 5))
incident.div(incident.sum(1).astype(float), axis = 0).plot(kind = 'bar',
                                                           stacked = True,
                                                           figsize = (8, 5),
                                                           color = colors)

plt.title('Incident Type vs Collision Type', fontsize = 20)
plt.legend()
plt.show()
```

***OBSERVATIONS FROM ABOVE:***
    1. for Incident_type- Single & multi Vehicle Collision:
        there are certain number of collisions are done like
        front,rear and side
    2. for Incident_type-vehicle theft some collision:
        are done we need to investigate since some data is missing
    3. for Incident_type- parked car:
        most of the collision are done in rear side compared to side and front
